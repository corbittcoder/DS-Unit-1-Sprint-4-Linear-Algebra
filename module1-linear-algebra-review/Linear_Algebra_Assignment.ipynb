## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D


vectors = np.array([[0, 0, 0, .5, .5, .5], 
                    [0, 0, 0, .2, .1, .0],
                    [0, 0, 0, .1, .3, .3]])

X, Y, Z, U, V, W = zip(*vectors)
print(X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
plt.show()
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1,1,1,8])
h = np.array([3,3,3,3])

print(np.linalg.norm(g))
print(np.linalg.norm(h))
```

g is longer. Having a single large number makes a much bigger difference on the norm than many medium numbers.

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.array([[12, 11],
              [7, 10]])
j = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])
print(np.linalg.norm(h))
print(np.linalg.norm(j))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
print(np.linalg.inv(h))
print(np.linalg.inv(j))
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
print(h.dot(np.linalg.inv(h)))
print(j.dot(np.linalg.inv(j)))
```

Yeah, minus some rounding errors they are the same (that is to say, mathematically they both equal the identity matrix.)
